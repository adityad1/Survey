```python
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset('tips')
anscombe = sns.load_dataset('anscombe')
```

Importing required libraries and getting our datasets 'tips' and 'anscombe'.

```python
import matplotlib.style as mplstyle
mplstyle.use(['bmh','fast','seaborn-dark-palette'])
```

Stylizing our plots.

```python
hist,ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Histogram of the total bill as well as a density plot')
```

We get a density plot as well as a histogram here.

```python
hist,ax = plt.subplots()
ax = sns.distplot(tips['total_bill'],kde=False)
ax.set_title('Histogram of the total bill')
ax.set_xlabel('Total bill')
ax.set_ylabel('Frequency')
```

We get only the histogram here. 

```python
hist,ax = plt.subplots()
ax = sns.distplot(tips['total_bill'],hist=False)
ax.set_title('Density plot of the total bill')
ax.set_xlabel('Total bill')
ax.set_ylabel('Probability')
```

Here, we get only the density plot.

```python
hist,ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug = True)
ax.set_title('Rug plot')
ax.set_xlabel('Total bill')
```

Rug plot which i need to explore.

```python
count,ax = plt.subplots()
ax = sns.countplot('day',data=tips)
ax.set_title('Count of days')
ax.set_xlabel('Day of the week')
ax.set_ylabel('Frequency')
```

Basically a bar graph plotting the count of days.

```python
scatter, ax = plt.subplots()
ax = sns.regplot(x = 'total_bill',y = 'tip', data = tips)
ax.set_title('Scatterplot of Total Bill and Tip', fontsize = 15)
ax.set_xlabel('Total bill')
ax.set_ylabel('Tips')
```

We get a scatter plot as well as a regression line. If we don't need the regression line, we would have to add fit_reg=False into parenthesis.

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip',fontsize=15, y=1.02)
```

Here, we get a joint plot with both the scatterplot and individual count plots.
y = 1.03 moves our text above the countplot to avoid overlapping.

```python
hexbin = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip',fontsize=15, y=1.04)
```

Here, we get a hexbin which is basically a scatter plot of the histograms of the xy data.

```python
kde, ax = plt.subplots()
ax = sns.kdeplot(data = tips['total_bill'], data2 = tips['tip'], shade = True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
kde_joint = sns.jointplot(x ='total_bill', y ='tip',data = tips, kind = 'kde')
kde_joint.fig.suptitle('Kernel Density Plot of Total Bill and Tip', fontsize = 15, y = 1.04)
kde_joint.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
```

Above, we get a density plot. Without shade = True, we would have dark boundaries instead of the shaded regions.

```python
bar, ax = plt.subplots()
ax = sns.barplot(x ='time', y ='total_bill', data = tips)
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

Barplot of average of groups of data points in the attribute tips.

```python
box, ax = plt.subplots()
ax = sns.boxplot(x = 'time', y = 'total_bill', data=tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

A boxplot gives us minimum, first quartile, median, third quartile, maximum, and also outliers based on the range between quartiles.

```python
violin, ax = plt.subplots()
ax = sns.violinplot(x = 'time', y = 'total_bill', data = tips)
ax.set_title('Violin plot of total bill by time of day', y = 1.02)
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

A violin plot is similar to a boxplot except that here we get a density representation of values of data.

```python
fig = sns.pairplot(tips)
```

Pairplot function plots all the possible relations between numerical data attributes from our dataset.
And also histograms between univariate attributes.

```python
fig = sns.PairGrid(tips)
fig = fig.map_upper(sns.distplot, hist = False)
fig = fig.map_lower(sns.kdeplot, shade = True)
fig = fig.map_diag(sns.distplot, kde = False)
```

We can customize what we need in the upper, lower, diagonal graphs individually.

```python
violin, ax = plt.subplots()
ax = sns.violinplot(x = 'time', y = 'total_bill', hue = 'sex', data = tips, split = True)
```

Using the split parameter gives us both male and female in the same violin with two different colors whereas setting it as false give us separate violin plots for male, female and lunch, dinner leaving us with 4 plots. It's default value is false.

```python
scatter = sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', fit_reg = False)
```

The hue parameter can be passed in many more functions.

```python
fig = sns.pairplot(tips, hue= 'sex')
```

This is another function where the hue function comes in handy.

```python
scatter = sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', fit_reg = False, scatter_kws = {'s':tips['size']*20})
```

Fitting more and more variables into a graph is always necessary. After colour, we use the size to depict a variable.

```python
scatter = sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', fit_reg = False, scatter_kws = {'s':tips['size']*20}, markers = ['x','o'])
```

It gets a bit messy if we use circles for both the values.

```python
anscombe_plot = sns.lmplot(x = 'x', y = 'y', data = anscombe, fit_reg = False, col = 'dataset', col_wrap = 2)
```

Our next variable creates itself an extra plot for a data point(4 in this case).
Without col_wrap, we would have all the plots congested into a single row.

```python
facet = sns.FacetGrid(tips, col ='time')
facet.map(sns.distplot, 'total_bill', rug = True)
```

lmplot had created its own figure and axes. Many other functions in seaborn do not create their own figures but only the axes.
This is the method to use the only axes creating functions.

```python
facet = sns.FacetGrid(tips, col ='day', hue ='sex')
facet = facet.map(plt.scatter, 'total_bill', 'tip')
facet = facet.add_legend()
```

Multivariate plots are also possible using the FacetGrid.

```python
facet = sns.FacetGrid(tips, col ='time', row ='smoker', hue ='sex')
facet.map(plt.scatter, 'total_bill', 'tip')
```

This way, we can use attributes in different axes using faceting.

```python
facet = sns.catplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips, row = 'smoker', col = 'time', kind = 'violin')
```

Catplot can be used for retrieving violin plots.
