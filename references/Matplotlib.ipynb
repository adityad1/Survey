```python
def recode_sex(sex):
    if sex == 'Female':
       return 0
    else:
       return 1
tips['sex_color'] = tips['sex'].apply(recode_sex)
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(x = tips['total_bill'], y = tips['tip'], s = tips['size'] * 15, c = tips['sex_color'], alpha = 0.85, cmap = 'flag')
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

In this scatterplot, we add sex as a third variable by means of color and a fourth variable as size in terms of the size of the co-ordinates.
Here, alpha controls the transparency of the co-ordinates which allows us to view through certain overlapped co-ordinates.

# Plots Prebuilt in Pandas Objects

```python
tips = sns.load_dataset('tips')
```

Dataset tips from seaborn.

```python
tips.head()
```

```python
tips.shape
```

```python
tips.describe()
```

Exploring the dataset.

```python
fig, ax = plt.subplots()
ax = tips['total_bill'].plot.hist()
fig, ax = plt.subplots()
ax = tips[['total_bill', 'tip']].plot.hist(alpha = 0.6, bins = 15, ax = ax)
```

Histrogram pre-built in pandas.

```python
fig, ax = plt.subplots()
ax = tips['tip'].plot.kde()
```

Density plot pre-built in pandas.

```python
fig, ax = plt.subplots()
ax = tips.plot.scatter(x='total_bill', y='tip', ax=ax)
```

Scatter plot pre-built in pandas.

```python
fig, ax = plt.subplots()
ax = tips.plot.hexbin(x='total_bill', y='tip', ax=ax, gridsize=10)
```

Hexbin pre-built in pandas.  
Here, gridsize specifies the number of hexagons on the x-axis.

```python
fig, ax = plt.subplots()
ax = tips.plot.box(ax=ax)
```

Boxplot pre-built in pandas.
